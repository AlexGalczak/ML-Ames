```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
from geopy import distance
%matplotlib inline
```

```python
clean_data = pd.read_csv('data/Ames_HousePrice_cleaned.csv')
```

```python
clean_data = clean_data[~clean_data['Longitude'].isna()]
```

```python
cl_y = clean_data['SalePrice']
cl_x = clean_data.loc[:,clean_data.columns!='SalePrice']

cl_x.columns
```

```python
cl_x['distFromCenter'] = cl_x.apply(lambda x: distance.distance((x.loc['Latitude'],x.loc['Longitude']), (42.0228613,-93.6202381)).miles, axis = 1)

cl_x
```

```python
cl_x_dummies = pd.get_dummies(cl_x, columns=['MSSubClass','MSZoning','LotShape','LandContour',\
        'LotConfig', 'Neighborhood', 'BldgType', 'HouseStyle','MasVnrType', 'ExterQual', \
        'ExterCond', 'Foundation', 'BsmtQual','BsmtCond', 'BsmtExposure', 'BsmtFinType1','BsmtFinType2',
       'HeatingQC','CentralAir','KitchenQual','GarageType','PavedDrive','SaleCondition','Electrical'],\
        prefix=['mssc','msz','ls','lc','lcf','nghbr','bt','hs','mvt','eq','ec','f','bq','bc','be','bf1','bf2',\
               'hq','ca','kq','gt','pd','sc','el'], prefix_sep='_',drop_first=True)
```

```python
print(cl_x_dummies.shape)
print(cl_y.shape)

cl_x_dummies = cl_x_dummies.fillna(0)
```

```python
x_train, x_test, y_train, y_test = train_test_split(cl_x_dummies, cl_y, test_size=0.3, random_state=247, shuffle=True)
```

```python
x_train.to_csv('x_train_dummies.csv',index=False)
y_train.to_csv('y_train_dummies.csv',index=False)
x_test.to_csv('x_test_dummies.csv',index=False)
y_test.to_csv('y_test_dummies.csv',index=False)
```
